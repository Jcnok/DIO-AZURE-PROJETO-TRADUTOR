```python
import requests
from bs4 import BeautifulSoup

def extract_text_from_url(url):
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        for script_or_stayle in soup(["script", "style"]):
            script_or_stayle.decompose()
        text = soup.get_text(separator='')
        #limpar texto
        linhas = (line.strip() for line in text.splitlines())
        parts = (phrase.strip() for line in linhas for phrase in line.split("  "))
        texto_limpo = '\n'.join(part for part in parts if part)
        return texto_limpo
    else:
        print(f"Failed to fetch the URL. Status code: {response.status_code}")
        return None

    soup = BeautifulSoup(response.text, 'html.parser')
    text = soup.get_text()
    return text

extract_text_from_url('https://dev.to/mrcaption49/collection-and-record-context-of-oracle-sql-50l0')
```
